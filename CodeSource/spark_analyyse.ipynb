```python
# Supprimer les lignes contenant des valeurs manquantes dans les colonnes spécifiées
patents = patents.na.drop(subset=["Inventor Name", "Title"])
```

```python
# Importer les fonctions de Spark SQL
from pyspark.sql.functions import col, count

# Compter les valeurs non nulles dans chaque colonne
missing_counts = patents.select([count(col(c)).alias(c) for c in patents.columns])

# Afficher les résultats
missing_counts.show()
```

```python
from pyspark.sql.functions import isnan, when, count, col

# Compter les valeurs NULL dans chaque colonne
null_counts = patents.select([count(when(isnan(c) | col(c).isNull(), c)).alias(c) for c in patents.columns])

# Afficher les résultats
null_counts.show()
```

```python
from pyspark.sql.functions import col, count, desc

# Analyse de la répartition des types de documents
document_type_distribution = patents.groupBy("Document Type").agg(count("*").alias("Count")).orderBy(desc("Count"))
document_type_distribution.show()
```

```python
# Analyse des inventeurs les plus prolifiques
prolific_inventors = patents.groupBy("Inventor Name").agg(count("*").alias("Count")).orderBy(desc("Count"))
prolific_inventors.show()
```

```python
# Sélectionner les colonnes "Filing Date" et "Publication Date"
dates_df = patents.select("Filing Date", "Publication Date")

# Afficher les résultats
dates_df.show(3500)
```

```python
dates = patents.select("Filing Date")
dates.show(3000)
```

```python
from pyspark.sql.functions import when, to_date

# Appliquer la conversion en fonction du format de date
patents = patents.withColumn("Filing Date", 
    when(patents["Filing Date"].rlike("^\\d{4}-\\d{2}-\\d{2}$"), 
         to_date(patents["Filing Date"], "yyyy-MM-dd"))
    .when(patents["Filing Date"].rlike("^\\d{2}/\\d{2}/\\d{4}$"), 
         to_date(patents["Filing Date"], "MM/dd/yyyy"))
    .otherwise(None))

# Afficher les résultats
patents.select("Filing Date").show(3500)
```

```python
from pyspark.sql.functions import year, month, dayofmonth, count

patents_non_null = patents.filter(patents["Filing Date"].isNotNull())

patents_dates_analysis = patents_non_null.withColumn("Year", year("Filing Date")) \
                                         .withColumn("Month", month("Filing Date")) \
                                         .withColumn("Day", dayofmonth("Filing Date"))

patents_yearly_count = patents_dates_analysis.groupBy("Year").agg(count("*").alias("Count")).orderBy("Year")

patents_yearly_count.show(100)
```

```python
patents = patents.withColumn("Publication Date", 
    when(patents["Publication Date"].rlike("^\\d{4}-\\d{2}-\\d{2}$"), 
         to_date(patents["Publication Date"], "yyyy-MM-dd"))
    .when(patents["Publication Date"].rlike("^\\d{2}/\\d{2}/\\d{4}$"), 
         to_date(patents["Publication Date"], "MM/dd/yyyy"))
    .otherwise(None))
```

```python
patents.select("Publication Date").show(3500)
```

```python
patents_non_null_1 = patents.filter(patents["Publication Date"].isNotNull())

patents_dates_analysis_1 = patents_non_null_1.withColumn("Year", year("Filing Date")) \
                                         .withColumn("Month", month("Filing Date")) \
                                         .withColumn("Day", dayofmonth("Filing Date"))

patents_yearly_count1 = patents_dates_analysis_1.groupBy("Year").agg(count("*").alias("Count")).orderBy("Year")

patents_yearly_count1.show(100)
```

# NLP

```python
titles = patents.select("Title")
titles.show(3000)
```

```python
from pyspark.sql.functions import split, explode, desc

words = titles.select(explode(split(titles.Title, " ")).alias("word"))
word_freq = words.groupBy("word").count().orderBy(desc("count"))
word_freq.show()

```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

word_list = [row['word'] for row in word_freq.collect()]
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(word_list))
plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from langdetect import detect
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def detect_lang(title):
    try:
        return detect(title)
    except:
        return "Unknown"

detect_lang_udf = udf(detect_lang, StringType())
```

```python
titles_with_lang = titles.withColumn("Language", detect_lang_udf(col("Title")))
```

```python
language_counts = titles_with_lang.groupBy("Language").count()
language_counts.show(100)
```

```python
lang = language_counts.toPandas()
lang.to_csv("langue.csv", index=False)
```
