```python
file_paths = [
    "C:/Users/hp/Downloads/Portfolio_1677394_04_19_2024 (1).xls",
    "C:/Users/hp/Downloads/Portfolio_1677394_04_19_2024 (2).xls",
    "C:/Users/hp/Downloads/Portfolio_1677394_04_19_2024 (3).xls"
]

# Liste pour stocker les DataFrames de chaque fichier
dfs = []

# Charger chaque fichier Excel dans un DataFrame et l'ajouter à la liste dfs
for file_path in file_paths:
    df = pd.read_excel(file_path)
    dfs.append(df)

# Fusionner les DataFrames en un seul
merged_df = pd.concat(dfs, ignore_index=True)

# Exporter le DataFrame fusionné au format Excel
merged_df.to_excel("merged_portfolio.xlsx", index=False)

print("Les fichiers ont été fusionnés avec succès.")
```

```python
non_english_count = 0

# Index des colonnes à vérifier
columns_to_check = ['Title', 'Abstract', 'Inventor Name', 'Assignee']

# Parcourir chaque ligne et vérifier si au moins une colonne n'est pas en anglais
for index, row in merged_df.iterrows():
    for column in columns_to_check:
        if pd.notnull(row[column]) and isinstance(row[column], str):
            try:
                if detect(row[column]) != 'en':
                    non_english_count += 1
                    break  # Sortir de la boucle interne dès qu'une détection non anglaise est trouvée
            except Exception as e:
                pass

print("Nombre de lignes qui ne sont pas en anglais :", non_english_count)
```

```python
lang_counts = {'unknown': 0}

# Parcourir chaque ligne et détecter la langue pour chaque colonne
for index, row in merged_df.iterrows():
    detected = False
    for column in merged_df.columns:
        if pd.notnull(row[column]) and isinstance(row[column], str):
            try:
                lang = detect(row[column])
                detected = True
                lang_counts.setdefault(lang, 0)
                lang_counts[lang] += 1
                break  # Sortir de la boucle interne dès qu'une détection de langue est réussie
            except Exception as e:
                pass
    if not detected:
        lang_counts['unknown'] += 1

# Afficher le nombre de lignes pour chaque langue
for lang, count in lang_counts.items():
    print(f"{lang.capitalize()}: {count} lignes")
```

```python
import os
from langdetect import detect
import pandas as pd

# Charger les données depuis le fichier Excel
translated_data = pd.read_csv(r"C:\\Users\\hp\\id2\\bg\\fpo_data.csv")

# Créer un dictionnaire pour stocker les DataFrames par langue
language_dataframes = {'unknown': pd.DataFrame()}

# Parcourir chaque ligne et détecter la langue pour le titre de chaque colonne
for index, row in translated_data.iterrows():
    title = row['Title']  # Supposons que 'Title' soit le nom de la colonne contenant les titres
    if pd.notnull(title) and isinstance(title, str):
        try:
            lang = detect(title)
            language_dataframes.setdefault(lang, pd.DataFrame())
            language_dataframes[lang] = pd.concat([language_dataframes[lang], pd.DataFrame(row).T])
        except Exception as e:
            language_dataframes['unknown'] = pd.concat([language_dataframes['unknown'], pd.DataFrame(row).T])

# Créer le répertoire de sortie s'il n'existe pas
output_directory = r"C:\\Users\\hp\\id2\\bg\\idiomas"
os.makedirs(output_directory, exist_ok=True)

# Enregistrer chaque DataFrame dans un fichier CSV dans le répertoire de sortie
for lang, df in language_dataframes.items():
    if not df.empty:
        output_file = os.path.join(output_directory, f"{lang}_titles.csv")
        df.to_csv(output_file, index=False)
```
